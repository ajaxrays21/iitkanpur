DataSet: https://bit.ly/2CudMVh
1. Plot BAR chart average of beer drinks consumption in continent.
2. Plot pie chart of percentage cunsumption of wine in continent, explde max percentage.
3. Plot the mean alcohol consumption per continent for every column.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
user=pd.read_csv("https://bit.ly/2CudMVh", delimiter=',')
print(user)
user1= user.groupby('continent').beer_servings.mean()
user1.plot.bar()
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
user=pd.read_csv("https://bit.ly/2CudMVh", delimiter=',')
print(user)

user2=user.groupby('continent').wine_servings.mean()
print(user2.argmax)
explode=(0,0,0.1,0,0)
user2.plot.pie(explode=explode, autopct='%1.1f%%')
```
